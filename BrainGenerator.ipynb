```python
class Transformations():
    def __init__(self, hyperparameters):
        self.hyperparameters = hyperparameters
        
    def transform_img(self, image):
        toTensor = transforms.ToTensor()
        image = toTensor(image)
        
        # Remove later
        test = image.squeeze(0)
        plt.imshow(test)
        plt.show()
        
        
        # Image dimensions: [N: batch size, C: # channels, H: height, W: width] 
        while (len(image.size()) < 4):
            image = image.unsqueeze(0)
        
        if "affine" in self.hyperparameters:
            image = self._affine_transform(image)
            
        return image
    
    def _affine_transform(self, image):
        affine_param = self.hyperparameters["affine"]
        affine_matrix = np.eye(2, dtype=np.float32)
        
        if "rotation" in affine_param:
            rot_param = affine_param["rotation"]
            rot_deg = np.deg2rad(np.random.uniform(rot_param[0], rot_param[1]))
            rot_matrix = np.array([[np.cos(rot_deg), np.sin(rot_deg)], [-np.sin(rot_deg), np.cos(rot_deg)]], dtype=np.float32)
            affine_matrix = np.dot(affine_matrix, rot_matrix)

        if "scaling" in affine_param:
            sc_param = affine_param["scaling"]
            sc_x, sc_y = np.random.uniform(sc_param[0], sc_param[1], 2)
            sc_matrix = np.array([[sc_x, 0], [0, sc_y]], dtype=np.float32)
            affine_matrix = np.dot(affine_matrix, sc_matrix)
        
        if "shearing" in affine_param:
            sh_param = affine_param["shearing"]
            sh_x, sh_y = np.random.uniform(sh_param[0], sh_param[1], 2)
            sh_matrix_x = np.array([[1, sh_x], [0, 1]], dtype=np.float32)
            sh_matrix_y = np.array([[1, 0], [sh_y, 1]], dtype=np.float32)
            affine_matrix = np.dot(affine_matrix, sh_matrix_x)
            affine_matrix = np.dot(affine_matrix, sh_matrix_y)
            
        affine_matrix = np.expand_dims(affine_matrix, axis=0)
        
        if "translation" in affine_param:
            tr_param = affine_param["translation"]
            tr_x, tr_y = np.random.uniform(tr_param[0], tr_param[1], 2)
            tr_vec = np.array([[[tr_x], [tr_y]]], dtype=np.float32)
        else:
            tr_vec = np.array([[[0], [0]]], dtype=np.float32)
        
        affine_matrix = np.concatenate((affine_matrix, tr_vec), axis=2)
        print("Affine matrix: ", affine_matrix)
        
        affine_matrix = torch.tensor(affine_matrix, dtype=torch.float32)
        grid = F.affine_grid(affine_matrix, image.size())
        transformed_image = F.grid_sample(image, grid)
        
        return transformed_image
   
```

```python
import random

def generateData(dataPath, hyperparamPath, batchSize=1):
    with open("hyperparameters.json", "r") as file:
        hyperparameters = json.load(file)
        
    transformations = Transformations(hyperparameters)
    
    # Assumes data files are {id}.generation.tif, where id = 1, ..., 100
    DATA_NUM = 100
    for i in range(batchSize):
        id = random.randint(1,DATA_NUM)
        label_map = Image.open(dataPath + '{:03d}'.format(id) + ".generation.tif")
        transformed_label_map = transformations.transform_img(label_map)
    # todo: extend to batchSize > 1
    return transformed_label_map
    
TRAINING_DIR = PATH_OF_DATA + 'training/'
label_maps = generateData(TRAINING_DIR, "./hyperparameters.json", batchSize=1)

for label_map in label_maps:
    label_map = label_map.squeeze(0)
    plt.imshow(label_map)
    plt.show()

```
